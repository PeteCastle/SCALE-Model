```python
import os
import shutil

# Store all file paths in a list
file_paths = []
for root, dirs, files in os.walk('D:\\Educational Others\\MosquitoDetectionSystemV2\\augmented_data\\test'):
    for name in files:
        file_paths.append(os.path.join(root, name))

# Function to split a list into chunks of n
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# Split the file paths into chunks of 24
file_groups = list(chunks(file_paths, 24))

# Define the base directory for the new directories
base_dir = 'D:\\Educational Others\\MosquitoDetectionSystemV2\\augmented_data\\grouped_test'

# Copy the files into separate directories for each group
for i, group in enumerate(file_groups):
    # Create a new directory for the group
    new_dir = os.path.join(base_dir, f'group_{i + 1}')
    os.makedirs(new_dir, exist_ok=True)
    
    # Copy the files into the new directory
    for file_path in group:
        shutil.copy2(file_path, new_dir)
```
