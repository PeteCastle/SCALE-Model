```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
FILE_PATH = '../../output/metrics.json'
record_df = pd.read_json(FILE_PATH, lines=True)
record_df.columns

record_df = record_df.rename(columns = {
    "cls_accuracy": "Classification Accuracy",
    "false_negative": "False Negative Rate",
    "fg_cls_accuracy": "Foreground Classification Accuracy",
    "iteration": "Iteration",
    "loss_box_reg": "Box Regression Loss",
    "loss_cls": "Classification Loss",
    "loss_rpn_cls": "Region Proposal Network Classification Loss",
    "loss_rpn_loc": "Region Proposal Network Localization Loss",
    "lr": "Learning Rate",
    "rank_data_time": "Data Loading Time",
    "roi_head/num_bg_samples": "Number of Background Samples",
    "roi_head/num_fg_samples": "Number of Foreground Samples",
    "rpn/num_neg_anchors": "Number of Negative Anchors",
    "rpn/num_pos_anchors": "Number of Positive Anchors",
    "time": "Time",
    "total_loss": "Total Loss",
    "data_time": "Data Loading Time",
    "eta_seconds": "ETA",
    "fast_rcnn/cls_accuracy": "Classification Accuracy",
    "fast_rcnn/false_negative": "False Negative Rate",
    "fast_rcnn/fg_cls_accuracy": "Foreground Classification Accuracy",
})
```

```python
metrics = [
    "Box Regression Loss",
    "Classification Loss",
    "Region Proposal Network Classification Loss",
    "Region Proposal Network Localization Loss",
    "Classification Accuracy",
    "False Negative Rate",
    "Foreground Classification Accuracy",
    "Total Loss"
]
r_epochs = len(record_df)  # assuming record_df is your DataFrame

fig, axs = plt.subplots(2, 4, figsize=(20,10))  # create a 4x2 grid of Axes
axs = axs.ravel()  # flatten the grid to easily iterate over it

for i, column in enumerate(metrics):
    axs[i].plot(np.arange(0, r_epochs), record_df[column], 'r')
    axs[i].set_title(column)
    axs[i].set_xlabel('Epochs')

plt.tight_layout()  # adjust the layout so that the plots do not overlap
plt.show()
```
